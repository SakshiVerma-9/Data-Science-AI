```python
import numpy as np

# Student Marks Dataset (out of 100)
marks = np.array([45, 67, 89, 90, 56, 78, 88, 92, 61, 74,
                  83, 95, 52, 68, 71, 85, 77, 64, 58, 81])

print("Dataset:")
print(marks)
```

```python
print( np.mean(marks))
```

```python
np.median(marks)
```

```python
# 3. Mode
values, counts = np.unique(marks, return_counts=True)
max_count = np.max(counts)

if max_count == 1:
    print("3. Mode: No mode (all values occur once)")
else:
    mode = values[counts == max_count]
    print("3. Mode:", mode)
```

```python
np.percentile(marks,75)
```

```python
# 11. Interquartile Range (IQR)
Q1 = np.percentile(marks, 25)
Q3 = np.percentile(marks, 75)
print("11. IQR:", Q3 - Q1)
```

```python
# 12. Students scoring above the mean
mean = np.mean(marks)
print( np.sum(marks > mean))
```

```python
# 13. Students scoring below 60
print(np.sum(marks < 60))
```

```python
# 14. Sum and Manual Average
total = np.sum(marks)
average = total / len(marks)
print(total)
print(average)
```

```python
# 15. Sorting and Top 5 Scores
ascending = np.sort(marks)
descending = np.sort(marks)[::-1]

print(ascending)
print(descending)
print(descending[:5])
```
