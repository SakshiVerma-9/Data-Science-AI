```python
#recursive function

def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
num=int(input("Enter a number whose factorial you want to calculate"))
f=factorial(num)
print(f"factorial of {num}={f}")
```

```python
def table(num, i=1):
    if i > 10:
        return
    
    print(num, "x", i, "=", num*i)
    
    table(num, i+1)   # recursive call


n = int(input("Enter any number: "))

table(n)
```

```python
#OOPS

class Student:
    Courses = "Data Science"

    def __init__(self, rn, n, c, p, ds):   # Constructor
        self.rollno = rn
        self.name = n
        self.cLan = c
        self.python = p
        self.dataScience = ds

    def displayStudentReport(self):
        tot = self.cLan + self.python + self.dataScience
        per = (tot / 300) * 100

        if per >= 80:
            grade = "A"
        elif per >= 60:
            grade = "B"
        elif per >= 40:
            grade = "C"
        else:
            grade = "D"

        print("\n------------ STUDENT REPORT CARD ------------")
        print("---------------------------------------------")
        print(f"Course      : {Student.Courses}")
        print(f"Name        : {self.name}")
        print(f"Roll No.    : {self.rollno}")
        print("---------------------------------------------")
        print(f"C Language  : {self.cLan}")
        print(f"Python      : {self.python}")
        print(f"DataScience : {self.dataScience}")
        print("---------------------------------------------")
        print(f"Total Marks : {tot}/300")
        print(f"Percentage  : {per:.2f}%")
        print(f"Grade       : {grade}")
        print("---------------------------------------------")


# Creating object
s1 = Student(102, "Sakshi Verma", 88, 85, 95)

# Calling method
s1.displayStudentReport()





    
   
        
    
```

```python
# Q2

class Employee:
    def __init__(self,Employee_ID,Name,Salary):
        self.Employee_ID= Employee_ID
        self.Name= Name
        self.Salary= Salary
    def display(self):
        print("Employee Information")
        print(f"Employee_ID :{self.Employee_ID}")
        print(f"Name        :{self.Name}")
        print(f"Salary      :{self.Salary}")
e1=Employee(101,"Sakshi",1000000)
e1.display()
```

```python
# Q3

class Bank_Account:
    def __init__(self,Account_No,Balance):
        self.Account_No= Account_No
        self.Balance=Balance
    def deposit(self,amount):
        self.Balance+=amount
        print(amount, "Deposited")
    def display_Balance(self):
        print(f"Account_No :{self.Account_No}")
        print(f"Updated_Balance:{self.Balance}")
#object creation
b1=Bank_Account(1230432549,500000)

#Deposite money
b1.deposit(5000)

#Display Upadated Balance
b1.display_Balance()
    
```
