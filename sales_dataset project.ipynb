```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
data = {
    "Brand": ["Toyota", "Honda", "Hyundai", "Tata", "Mahindra", "Kia"],
    "Sales": [120, 95, 110, 85, 100, 90],
    "Price": [12, 10, 11, 9, 14, 13],
    "Mileage": [18, 20, 19, 22, 17, 18]
}

df = pd.DataFrame(data)

# Create Dashboard
plt.figure(figsize=(12,8))

# Graph 1
plt.subplot(2,2,1)
sns.barplot(
    data=df,
    x="Brand",
    y="Sales",
    hue="Brand",
    palette="Set2",
    legend=False
)
plt.title("Car Sales")

# Graph 2
plt.subplot(2,2,2)
sns.scatterplot(
    data=df,
    x="Price",
    y="Sales",
    color="green",
    s=150
)
plt.title("Price vs Sales")

# Graph 3
plt.subplot(2,2,3)
sns.histplot(
    data=df,
    x="Mileage",
    bins=6,
    color="orange"
)
plt.title("Mileage Distribution")

# Graph 4
plt.subplot(2,2,4)
sns.boxplot(
    data=df,
    y="Sales",
    color="skyblue"
)
plt.title("Sales Outlier Detection")

# Display all graphs together
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    "Patient": ["P1","P2","P3","P4","P5","P6","P7","P8"],
    "Age": [25,40,35,60,45,30,55,50],
    "BloodPressure": [120,140,130,160,150,125,155,145],
    "Sugar": [90,110,105,180,150,95,170,140]
}

df = pd.DataFrame(data)
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(df["Patient"],
         df["BloodPressure"],
         marker="o",
         color="blue")
plt.title("Blood Pressure")
plt.xlabel("Patient")
plt.ylabel("Blood Pressure")
plt.subplot(2,2,2)
plt.scatter(df["Age"],
            df["Sugar"],
            color="red",
            s=120)
plt.title("Age vs Sugar")
plt.xlabel("Age")
plt.ylabel("Sugar")
plt.subplot(2,2,3)
plt.hist(df["BloodPressure"],# Graph 3: Histogram
         bins=5,
         color="purple")
plt.title("Blood Pressure Distribution")
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency")

# Graph 4: Box Plot
plt.subplot(2,2,4)     # Graph 4: Box Plot
plt.boxplot(df["Sugar"],
            patch_artist=True,
            boxprops=dict(facecolor="lightgreen"))
plt.title("Sugar Outliers")
plt.ylabel("Sugar")

# Adjust Layout
plt.tight_layout()

# Show Dashboard
plt.show()
```
