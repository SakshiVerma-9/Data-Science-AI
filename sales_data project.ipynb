```python
#(Q41- Q50)
# 41. Sales Representative generated highest revenue
df.groupby("Sales_Rep")["Sales_Amount"].sum().idxmax()

# 42. Region generated highest revenue
df.groupby("Region")["Sales_Amount"].sum().idxmax()

# 43. Product Category sold highest quantity
df.groupby("Product_Category")["Quantity_Sold"].sum().idxmax()

# 44. Customer Type generated highest revenue
df.groupby("Customer_Type")["Sales_Amount"].sum().idxmax()

# 45. Most used Payment Method
df["Payment_Method"].value_counts().idxmax()

# 46. Sales Channel with highest average sales
df.groupby("Sales_Channel")["Sales_Amount"].mean().idxmax()

# 47. Create Profit Column
df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]

# Display first five rows
df.head()

# 48. Top 10 Most Profitable Transactions
df.sort_values(by="Profit", ascending=False).head(10)

# 49. Monthly Sales Trend
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

monthly_sales = df.groupby(df["Sale_Date"].dt.month_name())["Sales_Amount"].sum()

monthly_sales
```

```python
# EDA Dashboard
import matplotlib.pyplot as plt

# Profit Column
df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]

# Monthly Sales
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])
monthly_sales = df.groupby(df["Sale_Date"].dt.month_name())["Sales_Amount"].sum()

plt.figure(figsize=(18,15))

# Total Sales
plt.subplot(3,3,1)
plt.bar(["Sales"], [df["Sales_Amount"].sum()])
plt.title("Total Sales")

# Total Profit
plt.subplot(3,3,2)
plt.bar(["Profit"], [df["Profit"].sum()])
plt.title("Total Profit")

# Region-wise Sales
plt.subplot(3,3,3)
df.groupby("Region")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Region-wise Sales")

# Category-wise Sales
plt.subplot(3,3,4)
df.groupby("Product_Category")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Category-wise Sales")

# Monthly Sales Trend
plt.subplot(3,3,5)
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")

# Sales by Sales Representative
plt.subplot(3,3,6)
df.groupby("Sales_Rep")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Sales by Sales Rep")

# Customer Type Distribution
plt.subplot(3,3,7)
df["Customer_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Customer Type")

# Payment Method Distribution
plt.subplot(3,3,8)
df["Payment_Method"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method")

# Sales Channel Distribution
plt.subplot(3,3,9)
df["Sales_Channel"].value_counts().plot(kind="bar")
plt.title("Sales Channel")

plt.tight_layout()
plt.show()

# Discount Analysis
plt.figure(figsize=(6,4))
plt.hist(df["Discount"], bins=10)
plt.title("Discount Analysis")
plt.xlabel("Discount")
plt.ylabel("Frequency")
plt.show()
```
