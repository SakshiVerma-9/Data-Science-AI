```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.DataFrame({
    "Order_ID": range(101, 121),
    "Product": [
        "Laptop","Mobile","Tablet","Laptop","Mobile",
        "Tablet","Laptop","Mobile","Tablet","Laptop",
        "Mobile","Tablet","Laptop","Mobile","Tablet",
        "Laptop","Mobile","Tablet","Laptop","Mobile"
    ],
    "Category": [
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics"
    ],
    "City": [
        "Delhi","Mumbai","Lucknow","Delhi","Lucknow",
        "Mumbai","Delhi","Lucknow","Mumbai","Delhi",
        "Lucknow","Mumbai","Delhi","Lucknow","Mumbai",
        "Delhi","Lucknow","Mumbai","Delhi","Lucknow"
    ],
    "Sales": [
        65000,30000,25000,70000,32000,
        27000,68000,35000,26000,72000,
        33000,28000,75000,34000,29000,
        76000,36000,30000,78000,37000
    ],
    "Profit": [
        12000,5000,4500,13000,5500,
        4700,12500,6000,4800,14000,
        6200,5000,14500,6300,5200,
        15000,6500,5400,15500,6800
    ],
    "Quantity": [
        5,10,8,6,9,
        7,5,11,8,6,
        10,7,5,9,8,
        4,12,9,5,10
    ]
})

print(df)
```

```python
df.head()
```

```python
import plotly.express as px
fig = px.scatter(df,
                 x="Sales",
                 y="Profit",
                 title="sales vs profit")
fig.show()
```

```python
# Q1
product_sales = df.groupby("Product")["Sales"].sum().reset_index()

fig = px.bar(
    product_sales,
    x="Product",
    y="Sales",
    title="Total Sales by Product",
    color="Product"
)

fig.show()

```

```python
# Q2
#product_sales = df.groupby("Product")["Sales"].sum().reset_index()

fig = px.pie(
    product_sales,
    names="Product",
    values="Sales",
    title="Sales Distribution by Product",
   
)

fig.show()
```

```python
# Q3
fig = px.scatter(df,
                 x="Sales",
                 y="Profit",
                 size="Quantity",
                 color="Product",
                 title="sales vs profit"
)
fig.show()
```

```python
# Q4
fig=px.histogram(
    df,
    x="Sales",
    nbins=10,
    title="Histogram of Sales "
  
)
fig.show()
```

```python
# Q5
fig=px.box(
    df,
    x="Product",
    y="Sales",
    color="Product",
    title="Sales distribution of each product "
   
  
)
fig.show()
```

```python
# Q6
product_sales = df.groupby("City")["Sales"].sum().reset_index()

fig = px.bar(
    product_sales,
    x="City",
    y="Sales",
    title="Total Sales of each city",
    color="City"
)

fig.show()
```

```python
# Q7
avg_profit = df.groupby("Product",as_index=False )["Profit"].mean()

fig = px.line(
    avg_profit,
    x="Product",
    y="Profit",
    title="Average profit of each product ",
    
)

fig.show()
```
