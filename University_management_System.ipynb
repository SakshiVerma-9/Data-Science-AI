```python
# University Management System
class University:
    def __init__(self, university_name, location):
        self.university_name = university_name
        self.location = location

    def display_university(self):
        print("\n----- UNIVERSITY DETAILS -----")
        print("University Name :", self.university_name)
        print("Location        :", self.location)


class Department(University):
    def __init__(self, university_name, location, department_name, hod):
        super().__init__(university_name, location)
        self.department_name = department_name
        self.hod = hod

    def display_department(self):
        self.display_university()
        print("\n----- DEPARTMENT DETAILS -----")
        print("Department Name :", self.department_name)
        print("HOD             :", self.hod)


class Student(Department):
    def __init__(self, university_name, location, department_name, hod,
                 student_name, roll_no, semester, marks):
        super().__init__(university_name, location, department_name, hod)
        self.student_name = student_name
        self.roll_no = roll_no
        self.semester = semester
        self.marks = marks

    def display_student(self):
        self.display_department()
        print("\n----- STUDENT DETAILS -----")
        print("Student Name :", self.student_name)
        print("Roll No      :", self.roll_no)
        print("Semester     :", self.semester)
        print("Marks        :", self.marks)
        print("Result       :", "Pass" if self.marks >= 40 else "Fail")


class Faculty(Department):
    def __init__(self, university_name, location, department_name, hod,
                 faculty_name, faculty_id, subject):
        super().__init__(university_name, location, department_name, hod)
        self.faculty_name = faculty_name
        self.faculty_id = faculty_id
        self.subject = subject

    def display_faculty(self):
        self.display_department()
        print("\n----- FACULTY DETAILS -----")
        print("Faculty Name :", self.faculty_name)
        print("Faculty ID   :", self.faculty_id)
        print("Subject      :", self.subject)


# ---------------- PROGRAM ----------------

print("====== UNIVERSITY MANAGEMENT SYSTEM ======\n")

# University Details
uname = input("Enter University Name: ")
loc = input("Enter University Location: ")

# Department Details
dept = input("Enter Department Name: ")
hod = input("Enter HOD Name: ")

# Student Details
print("\nEnter Student Details")
sname = input("Student Name: ")
roll = input("Roll No: ")
sem = input("Semester: ")
marks = float(input("Marks: "))

student = Student(uname, loc, dept, hod, sname, roll, sem, marks)

# Faculty Details
print("\nEnter Faculty Details")
fname = input("Faculty Name: ")
fid = input("Faculty ID: ")
subject = input("Subject: ")

faculty = Faculty(uname, loc, dept, hod, fname, fid, subject)

# Display Data
print("\n========================================")
student.display_student()

print("\n========================================")
faculty.display_faculty()
     
       
            
            
        
```
