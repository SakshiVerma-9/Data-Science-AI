```python
import numpy as np

# Matrix A
A = np.array([
    [2, 4, 6],
    [8, 10, 12],
    [14, 16, 18]
])

# Matrix B
B = np.array([
    [1, 3, 5],
    [7, 9, 11],
    [13, 15, 17]
])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
```

```python
# Q1
print(A.shape)
print(A.size)
print(A.ndim)
print(B.shape)
print(B.size)
print(B.ndim)
```

```python
# Q2
print(A+B)
```

```python
# Q3
print(A-B)
```

```python
# Q4
print(A*B)
```

```python
# Q5
print(np.matmul(A,B))
```

```python
# Q6
print(A.T)
print(B.T)
```

```python
# Q7
print(np.linalg.det(A))
```

```python
# Q8
print(np.linalg.inv(A))
```

```python
# Q9
print(np.trace(A))
print(np.trace(B))
```

```python
# Q 10
print(np.max(A))
print(np.min(A))
print(np.mean(A))
print(np.sum(A))
```

```python
# Q11
print(np.sum(B,axis=1))
print(np.sum(B,axis=0))
```

```python
# Q12
print(A[0])
print(A[:,1])
print(A[-1])
print(A[-1,-1])
```

```python
# Q13
print(A.reshape(1,9))
print(B.flatten())
```

```python
# Q14
eigenvalues,eigenvectors=np.linalg.eig(A)
print(eigenvalues)
print(eigenvectors)
```
