```python
import numpy as np
arr= np.array([4,6,8,9])
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[[1,2,3],[4,5,6]],[[11,12,13],[14,15,16]]])
print(arr)
print(arr1)                          #array attributes: (ndim, shape, size, dtype)
print(arr1[1,0])
print(arr2)
print(arr.ndim)
print(arr1.ndim)
print(arr.size)
print(arr1.size)
print(arr2.size)
print(arr.shape)
print(arr1.shape)

print(arr2.shape)
print(arr.dtype)
print(arr1.dtype)

print(arr[0])
print(arr[1:3:1])






```

```python
arr = np.array([1,2,3,2,0], dtype= bool)

if np.array_equal(arr , arr[::-1]):
    print("Palindrome")
else:
    print("not palindrome")

```

```python
myarr=np.zeros([2,3])
print(myarr)
myarr1=np.ones((2,3))
print(myarr1)
myarr2=np.arange(1,10,2)
print(myarr2)
print(np.eye(3))
```

```python
#aggregate function: sum,min,max,dtype

arr=[2,5,7,8,9]
print(np.sum(arr))
print(np.min(arr))
print(np.max(arr))
print(np.dtype)
print("Average:", np.mean(arr))



```

```python
arr=[5,4,3,2,6,1]
print(np.sqrt(arr))
print(np.square(arr))
print(np.abs(arr))
print(np.exp(arr))
print(np.log(arr))
```
